```python
import cv2
from ultralytics import YOLO

def run_yolo_webcam():
    # Load YOLO model
    model = YOLO('main_model_v2.pt')  # Chn model YOLOv8 ph hp (e.g., yolov8n.pt, yolov8s.pt)

    # Khi to webcam (ID 0 l webcam mc nh)
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Cannot open webcam")
        return

    try:
        while True:
            # c frame t webcam
            ret, frame = cap.read()
            if not ret:
                print("Failed to grab frame")
                break

            # Chy m hnh YOLO trn frame
            results = model(frame)

            # Hin th kt qu
            annotated_frame = results[0].plot()  # Annotated frame vi bounding boxes v labels
            cv2.imshow('YOLOv8 Webcam', annotated_frame)

            # Thot nu nhn phm 'q'
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        # Gii phng ti nguyn
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    run_yolo_webcam()
```
