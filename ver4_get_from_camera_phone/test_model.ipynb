```python
import cv2
from ultralytics import YOLO

def run_yolo_webcam():
    # Load YOLO model
    model = YOLO('main_model_v2.pt')  # Chọn model YOLOv8 phù hợp (e.g., yolov8n.pt, yolov8s.pt)

    # Khởi tạo webcam (ID 0 là webcam mặc định)
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Cannot open webcam")
        return

    try:
        while True:
            # Đọc frame từ webcam
            ret, frame = cap.read()
            if not ret:
                print("Failed to grab frame")
                break

            # Chạy mô hình YOLO trên frame
            results = model(frame)

            # Hiển thị kết quả
            annotated_frame = results[0].plot()  # Annotated frame với bounding boxes và labels
            cv2.imshow('YOLOv8 Webcam', annotated_frame)

            # Thoát nếu nhấn phím 'q'
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        # Giải phóng tài nguyên
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    run_yolo_webcam()
```
